```python
import os
import pandas as pd
import numpy as np
import spacy
import re
from spacy.matcher import Matcher
nlp = spacy.load("en_core_web_sm")
matcher = Matcher(nlp.vocab)
pattern = [{"ENT_TYPE":{"IN":["PERSON","PRODUCT","GPE"]}}, {"ORTH": ","} , {"ENT_TYPE": "PERSON"}]
matcher.add("LastCommaFirst", None, pattern)
#pattern = [{"ENT_TYPE":"PERSON"},{"ENT_TYPE": "PERSON"}]
#matcher.add("FirstLast", None, pattern)
```

```python
text_flist = list()
for root, directories, filenames in os.walk('../../../../eis_documents/enepa_repository/text_as_datatable/'): 
    #for directory in directories: 
    #    print(os.path.join(root, directory))  
    for filename in filenames:  
        text_flist.append(os.path.join(root,filename)) 
```

```python
text_storage = '../../../input/filtered_text_files/'
flist = os.listdir(text_storage)
projects_used = pd.read_csv('../../../scratch/boilerplate/project_candidates_eis_only.csv')
files_used = pd.read_csv('../../../scratch/boilerplate/document_candidates_eis_only.csv')
file_names = [re.sub('\.pdf$','.txt',f) for f in files_used['FILE_NAME']]

full_text_to_search = pd.DataFrame()
```

```python
for file in file_names:
    print(file)
    file_name_match = [bool(re.findall(file,f)) for f in text_flist]
    if sum(file_name_match)>0:
        #file_name = list(filter(None, file_name_match))[0]
        file_index_loc = [i for i, x in enumerate(file_name_match) if x]
        text_table = pd.read_table(text_flist[file_index_loc[0]])
        good_match = [re.findall('Interdisciplinary Team|PREPARER\W|Preparer\W|Preparers\W|Contributors|PREPARERS|CONTRIBUTORS|Team Members',t) for t in text_table['text']]
        loc = [i for i, x in enumerate([i != list() for i in good_match]) if x]
        if(len(loc)!=0):
            bad_match = [re.findall('\.{7,}|TABLE OF CONTENTS|Table of Contents|WORKS CITED|BIBLIOGRAPHY|Works Cited|Bibliography',t) for t in text_table['text']]
            bad_loc = [i for i, x in enumerate([i != list() for i in bad_match]) if x]
            filter = [(l in bad_loc) == False for l in loc]
            filtered_locs = [i for indx,i in enumerate(loc) if filter[indx] == True]
            
            if(len(filtered_locs)!=0):
                added_loc = []
                for f in filtered_locs:
                    added_loc.append([f + r for r in range(1,2)])
                full_list = np.asarray([item for sublist in added_loc for item in sublist])
                full_list = full_list[full_list <= text_table.shape[0]]
                full_list = full_list[full_list >= 0]
                full_list = np.unique(full_list)
                full_list = full_list - 1
                #print(full_list)
                text_to_search = pd.DataFrame({'text':text_table['text'][full_list],'FILE':file,'PROJECT_ID':re.sub('_.*','',file)})
                full_text_to_search = full_text_to_search.append(text_to_search)
```

```python
full_text_to_search.index = range(full_text_to_search.index.shape[0])
```

```python
empty_entity = pd.DataFrame()
matcher_table = []

for i in range(full_text_to_search.shape[0]):
    tx = full_text_to_search['text'][i]
    pid = full_text_to_search['PROJECT_ID'][i]
    file = full_text_to_search['FILE'][i]
    #print(i)
    #print(pid)
    #print(file)
    doc = nlp(str(tx))
    table = []
    for ent in doc.ents:
        table.append([ent.text,ent.label_,spacy.explain(ent.label_)])
    df = pd.DataFrame(table, columns=['Entity', 'Label','Label_Description']).sort_values(by=['Label'])
    empty_entity = empty_entity.append(df)
    matches = matcher(doc)
    for match_id, start, end in matches:
        string_id = nlp.vocab.strings[match_id]  # Get string representation
        span = doc[start:end]  # The matched span
        matcher_table.append([pid,file,match_id, string_id, start, end, span.text])
        
```
